```python
# greatest of three numbers
a = int(input('Enter A : '))
b = int(input('Enter B : '))
c = int(input('Enter C : '))

if (a>b and a>c):
  print('A is bigger')
if(b>a and b>c):
  print('B is bigger')
if(c>a and c>b):
  print('C is bigger')
```
